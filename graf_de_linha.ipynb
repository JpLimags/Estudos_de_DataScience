### Gráfico de linhas

- Pode ser descrito como um tipo de gráfico que a infomração em forma de um conjunto de pontos (conhecido como markers) ligados por uma linha.

 # IMAGEM

- Gráficos de linhas podem ser utilizados , quando temos dados que se conectam e tem uma continuidade. Não sendo indicado em dados de tipo categórico;

- Ademais, os gráficos de linha são idéais para mostrar alterações e tendências de dados em intervalos iguais, sendo amplamente utilizados em séries históricas por exemplo;

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv', index_col='data', parse_dates=True)
df.head()
```

```python
#Gráfico que mostra o desenvolvimento dos casos de dengue dos anos de 1998 a 2014

fig, ax = plt.subplots(figsize=(8,5))

df["casos-confirmados"].plot(kind="line", ax=ax)
ax.set_title("Dengue em Campinas")
ax.set_ylabel("Casos Confirmados")
ax.set_xlabel("Data")

plt.tight_layout()
```

#### Trabalhando com intervalos

- Supondo que a temperatura influência no número de casos vamos analizar os dados da temperatura média no intervalo dos anos de 2011 á 2014

```python
#Supondo que a temperatura influência no número de casos vamos analizar os dados de 2011 á 2014

fig, ax = plt.subplots(figsize=(8,5))

df.loc["2011-01-01":"2014-12-31","temperatura-media"].plot(kind="line", ax=ax)
ax.set_title("Gráfico da temperatura média")
ax.set_ylabel("Temperatura")
ax.set_xlabel("Data")

plt.tight_layout()
```

#### Plotar múltiplos gráficos separados

```python
#Gráfico sobre as temperaturas

fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,4))

df.loc["2011-01-01":"2014-12-31", "temperatura-media"].plot(kind='line', ax=ax0)
ax0.set_title("Gráfico da temperatura média")
ax0.set_ylabel("Temperatura")
ax0.set_xlabel("Data")

#Gráfico sobre o número de casos no mesmo peíodo

df.loc["2011-01-01":"2014-12-31", "casos-confirmados"].plot(kind='line', ax=ax1)
ax1.set_title("Gráfico de casos confirmados")
ax1.set_ylabel("Casos confirmados")
ax1.set_xlabel("Data")

plt.tight_layout()
```

#### Plotar múltiplos gráficos sobrepostos

- Aqui se faz necessário a criação de um novo dataframe apartir dos antigo para normalização dos valores do nosso dados 

```python
# Normalização dos dados
df_norm = (df - df.mean()) / df.std()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,4))

#Plotagem da temperatura média
df_norm.loc["2011-01-01":"2014-12-31", "temperatura-media"].plot(kind="line", ax=ax)

#Plotagem dos casos confirmados
df_norm.loc["2011-01-01":"2014-12-31", "casos-confirmados"].plot(kind="line", ax=ax)

# plotagem dos titulos e do eixo

ax.set_title("Temperatura média X Casos Confirmados")
ax.set_xlabel("Data")

plt.tight_layout()
```
