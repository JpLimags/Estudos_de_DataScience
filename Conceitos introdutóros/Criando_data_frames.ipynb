## Criando DataFrames

  - Todos nós Sabemos que uma das estruturas base por trás do Pandas pode ser descrita como DataFrames, a ideias é que a partir deste notebook é aprendermos a criar DataFrames de diferentes formas.


  - Data frames podem ser descritos como uma tabela retangular(semelhante a que vemos no Excel), contendo uma coleção ordenada de colunas, em que cada uma pode ter um tipo de valor diferente(booleano, date, int...).


  - Outra informação relevante é que os DataFrames tem indice tanto para linha quanto para coluna; podendo ser imaginado como um dicionário de Series.

```python
#Importando o pandas

import pandas as pd
```

## Dataframe apartir de um dicionário do Python

```python
#Criando um dic

dados = {
    
    'nome': ['Carlos', 'Marcos', 'Raquel', 'Renata'], 
    'idade': [35, 43, 27, 19],
    'cidade': ['Araraquara', 'Gramado', 'Manaus', 'Fortaleza'],
    'comprou': [True, False, False, True]
}
```

```python
#Criando o dataframe

df = pd.DataFrame(dados)
df
```

##### Após a criação do nosso DataFrame também podemos realizar modificações, como por exemplo adicionar elementos ao mesmo. Como por exemplo, criar uma lista com id de usuários em nosso DF e passar essa lis como indíce do nosso DataFrame

```python
#Criando os id de usuários

id_usuario = [8559, 7982, 8890, 6780]
```

Para definirmos o index de um dataframe, utilizaremos a função `df.index`.

```python
df.index = id_usuario
df
```

## DataFrames a partir de Listas

```python
#Lista com os dados

dados = [['Carlos', 35, 'Araraquara', 'True'],
         ['Theo', 32, 'Natal', 'False'],
         ['Raquel', 12, 'SJC', 'True'],
         ['Fernanda', 14, 'Brasilia', 'True']]

#Criando o Dataframe

df = pd.DataFrame(dados, columns=['Nome', 'Idade', 'Cidade', 'Comprou'], index = [8559, 7982, 8890, 6780])
df
```

Caso seja necessário selecionar um determina linha por seu index poderemos utilizar a função `loc`.

```python
#Selecionando pelo index

df.loc[6780]
```

```python
#Criando uma nova coluna

df['saldo'] = 0.0

#Vizualizando DataFrame
df
```

### Modificando index e colunas

 - As vezes é ncessário alterar o nome dos índices (ou label das colunas).


```python
#Alterando o index

df.index = ['a','b','c','d']

#Printando o DataFrame
df
```

```python
#assim como fizemos paras as linhas podemos também fazer para colunas

df.columns = ['Nome do cliente', 'Idade', 'Naturalidade', 'Realizou Compra', 'Saldo na Loja']

df
```
