# Filtrando valores

## loc

- ##### Se quisermos acessar e filtrar valores de um DataFrame, podemos utilizar para filtrar os valores em um dataframe a função `loc[]`. A função loc basicamente acessa um grupo de colunas e linhas por uma label ou por uma matriz boleano. 

- Para usar esta opção, você deve entender que o mesmo funciona da seguinte maneira:

  `loc[ <linhas>, <colunas>]`

O loc pode ser utilizado em duas situações:

  - Selecionar linhas por *labels/index*
  - Selecionar linhas baseado em um *array* booleano.

Vejamos o exemplo abaixo: 

- Suponhamos que eu precise extrair a latitude e longitude do local onde aconteceu a ocorrência `201311259977425`

```python
#Filtrando valores com a função loc

df.loc[201311259977425, ['ocorrencia_latitude', 'ocorrencia_longitude']]
```

Outro exemplo, digamos que desejamos saber quais vôos eu tive a ocorrência do tipo "Falha do motor em voo":

```python
# Ver apenas as 5 primeiras linhas

df.loc [df['ocorrencia_tipo'] == "FALHA DO MOTOR EM VOO"]
```

## iloc


- ##### Se quisermos acessar e filtrar valores de um DataFrame, outra função  que podemos utilizar para filtrar os valores em um dataframe é a função `iloc[]` que é até mais simples que a loc. A função iloc faz o acesso através de uma indexação puramente baseada em localização inteira para seleção por posição. 

- Para usar esta opção, você deve entender que o mesmo funciona da seguinte maneira:

  `iloc[ <numero_linha>, <numero_coluna>]`

-  Nessa função podemos selecionar uma rande de interessE. Por exemplo 

```python
df.iloc[5:18]
```
