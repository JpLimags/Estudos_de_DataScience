# Visualização de dados com Pandas

- Muitas vezes quando estamos trabalhando com dados, fica difícil perceber relações entre eles ou mesmo parões visuais, já que nada mais são que um coglomerado de dados em forma de tabela, e é buscando resolver esse problem a que entra a Visualização de dados.

- A visualização de dados busca passar ao cientista uma visão antes não percebida, através de respresentações com inumeros gráficos diferentes,buscando facilitar a percepção de padrões e/ou entender aorganização dos mesmos.

```python
# Importando a biblioteca

import pandas as pd
```

```python
# Importando os dados

df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/BBAS3.SA.csv', parse_dates=True, index_col='Date')
df.head()
```

## Gráfico de linhas 

- Gráficos de linhas são muito utilizados para vermos o comportamento dos dados em séries temporais, por exemplo, se quisessemos saber saber quando foi o pico da covid-19 no Brasil poderiamos utilizar gráficos de barras para fazer tal visualização.

```python
df.High.plot()
```

## Histograma


- Um histograma pode ser descrito como sendo um tipode plotagem de barras que oferece uma exibição discreta das frquências dos valores. Onde os pontos são separados em compartimentos discretos descritos como **bins**.

```python
df['Volume'].hist()
```

## Gráfico de dispersão

-  Os gráficos de dispersão ou plotagens de pontos, são uma maneira coveniente de analisar o relacionacamento entre duas series de dados unidimensionais.

```python
#Importando o conjunto de dados sobre o consumo de cerveja


df_cerv = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/consumo_cerveja.csv")
df_cerv.head()
```

```python
#Plotando um gráfico de dispersão
#Vale ressaltar que precisamos de duas variáevi para fazermos a plotagem caso fassamos a a plotagem sem passar as variaveis neste caso, irá gerar um erro


df_cerv.plot.scatter('media','consumo')
```

## Bloxpot

- Boxplot é um tipo de gráfico hiper infromativo trazendo inúmeras informações estatísticas trazendo um noção mais precisa do dados.

image.png

- As informações estatísticas são os três quartis(25%,50%,75%), maximos e minimos mediana e outliers.

```python
#Plotando o boxplot

df_cerv.consumo.plot(kind='box')
```
